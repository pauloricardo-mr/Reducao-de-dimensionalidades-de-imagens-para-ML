##**2. Imagens**

> Essa secção possiblita o carregamento e análise das imagens que terão sua dimensionalidade reduzida. No exemplo, é usada uma imagem da atriz Lena conhecida por ser a modelo da primeira imagem a ser escaneada em um computador.

```python
img = cv.imread('/content/Lenna_(test_image).png')
cv2_imshow(img)
```

```python
plt.hist(img.ravel(),bins = 256, range = [0,256]) 
plt.show()
```

##**3. Mudança de dimensionalidade**

> Essa secção muda a dimensão da imagem de colorida para grayscale e então para binário(black&white).

```python
img=cv.cvtColor(img, cv.COLOR_BGR2GRAY) #Converte para GrayScale
cv2_imshow(img)
```

```python
(T,bin) = cv.threshold(img,135,255,cv.THRESH_BINARY) 
(T,binI) = cv.threshold(img,135,255,cv.THRESH_BINARY_INV) 
resultado = np.vstack(
    [np.hstack([img,bin]),
     np.hstack([binI,cv.bitwise_and(img, img, mask=binI)])])
cv2_imshow(resultado)
cv.waitKey(0)
```
